```python
import json

all_results = {
    "random baseline": "061826_random_results.json",
    "SIRL": "061826_sirl_results.json",
    "PCA": "061826_pca_results.json",
    "tversky-SIRL": "061826_tversky_4_results.json",
}
```

```python
import json
import plotly.graph_objects as go

means, stds = [], []
for path in all_results.values():
    with open(path) as f:
        d = json.load(f)
    if "mean" not in d:
        d = d["1000"]
    means.append(d["mean"])
    stds.append(d["std"])

fig = go.Figure(go.Bar(
    x=list(all_results),
    y=means,
    error_y=dict(type="data", array=stds),
))
fig.update_layout(title="fpe on 1000 pooled human queries", yaxis_title="fpe", template="plotly_white")
config = {
  'toImageButtonOptions': {
    'format': 'png', # or jpeg, svg, webp
    'scale': 4 # Increase this for higher resolution
  }
}
fig.show(config=config)
```
